```python
import numpy as np
import pandas as pd
import yfinance as yfin
import matplotlib.pyplot as plt

#velas japonesas
import mplfinance as mpf
```

```python
#Dascargar datos de CX Y EX
cx = yfin. download("CX", start="2020-01-01", end="2021-01-01")
ec = yfin.download("EC", start="2020-01-01", end="2021-01-01")
```

```python
#Grafico de velas para CX
mpf.plot(cx, type='candle', volume=True, figratio=(12,7), style='yahoo', title='CEMEX') 
```

```python
#Grafico de velas para EC
mpf.plot(ec["2020-01-01":"2020-04-01"],
         type='candle',
         volume=True,
         figratio=(12,7), 
         style='yahoo', 
         title='ECOPETROL'
) 
```

```python
#Dascargar datos de EC intervalo de una hora
ec2 = yfin. download("EC", start="2023-05-01", end="2023-05-05", interval="1h")
```

```python
#Grafico de velas junto a medias moviles de 5 y 20 horas
mpf.plot(ec2, type='candle', volume=True, figratio=(12,7), style='yahoo', title='ECOPETROL HORAS', mav=(5, 20)) 
```
