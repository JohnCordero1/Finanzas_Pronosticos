```python
#funciones numericas adicionales
import numpy as np

import pandas as pd
import yfinance as yfin
import matplotlib.pyplot as plt
```

```python
df =yfin.download("AAPL", start="2019-01-01", end="2021-01-01")
df
```

```python
#Retornos aritmeticos
df["Adj Close t-1"] = df["Adj Close"].shift(1) #shift desplaza filas
df["Retornos"] = (df["Adj Close"] - df["Adj Close t-1"])/df["Adj Close t-1"]
#Retornos logaritmicos
df["Log Returns"] = np.log(df["Adj Close"]) - np.log(df["Adj Close t-1"]) 
df
```

```python
#Graficar retornos
plt.figure(figsize=(10.8, 5.8))
plt.plot(df["Log Returns"], color = "red")
plt.title("Retornos logaritmicos")
plt.xlabel("Fecha")
plt.show()
```

```python
#volatilidad diaria
vol_d = np.std(df["Log Returns"])

#Calculamos la volatilidad diaria anualizada (252 dias bursatiles)
vol_a = vol_d*np.sqrt(252)
vol_a
```

```python
#Datos del S&P500
df2 =yfin.download("^GSPC", start="2015-01-01", end="2021-01-01")
df2
```
