```python
import yfinance as yfin
import pandas as pd

from google.colab import files

import matplotlib.pyplot as plt
```

```python
ticker = "GOOG"
df =yfin.download(ticker, start="2019-01-01", end="2021-01-01")
df
```

```python
ticker = "MSFT"
df2 =yfin.download(ticker, start="2019-01-01", end="2021-01-01")
df2
```

```python
#Estadisticas descriptivas de GOOGLE
df.describe()
```

```python
#Estadisticas descriptivas de MICROSOFT
df2.describe()
```

```python
#Grafico del volumen de MICROSOFT
plt.figure(figsize=(10.8, 4.8))
plt.plot(df2["Volume"], color = "blue")
plt.title("Microsft [MSFT]")
plt.xlabel("Fecha")
plt.ylabel("Volumen de transacción")
plt.show()
```

```python
#Grafico del de precios de cierre de MICROSOFT
plt.figure(figsize=(10.8, 4.8))
plt.plot(df2["Close"], color = "red")
plt.title("Microsft [MSFT]")
plt.xlabel("Fecha")
plt.ylabel("Precio de cierre")
plt.show()
```
