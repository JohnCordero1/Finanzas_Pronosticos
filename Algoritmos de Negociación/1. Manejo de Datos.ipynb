```python
#Api datos
import yfinance as yfin

#Manejo de tablas de datos
import pandas as pd

#Manejo de archivos, por si queremos bajar archvos de google colab
from google.colab import files

#Gráficos
import matplotlib.pyplot as plt
```

```python
from yfinance.multi import download
ticker = "AAPL"
df = yfin.download(ticker, start="2019-01-01", end="2021-01-01")
df #imprimir la variable
```

```python
#Puede que la tabla sea grande o queramos bajarlo a excel, por lo que guardamos en un archivo csv
#Guardar en csv
df.to_csv("AAPL.csv") #archivo guardado en google colab

#Una vez importado files, descargamos
files.download("AAPL.csv")
```

```python
#Cargar archivo
files.upload()
```

```python
#Ahora si importamos el archivo csv a nuestro entorno de programcion
df2 = pd.read_csv("AAPL.csv")
df2["Date"] = pd.to_datetime(df2["Date"]) #Establecemos columan de Date como formate de fecha
df2.set_index("Date", inplace=True) #set_index para colocar las fechas como nombres de las filas
df2
```

```python
#Calculamos estadisticas descriptivas
df2.describe()
```

```python
#Grafico de los precios de cierre de Apple, instalamos libreria matplotbit
plt.figure(figsize=(10.8, 4.8)) #tamaño de la grafica
plt.plot(df2["Close"], color = "blue")
plt.title("Apple [$AAPL]")
plt.xlabel("Fecha")
plt.ylabel("Precio de Cierre")
plt.show()
```
